# A program that counts the amount of GBP change out of a picture of coins

```python
import cv2
import numpy as np
from PIL import Image, ImageFilter, ImageFont, ImageDraw
import matplotlib.pyplot as plt
%matplotlib inline
```

## The picture

```python
file_name = 'capstone_coins.PNG'
im = Image.open(file_name)
plt.imshow(im)
```

## First steps

### Converting to gray

```python
image = cv2.imread('capstone_coins.PNG')
output = image.copy()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray=cv2.medianBlur(gray,5)
```

### Detecting circles

```python
# detect circles in the image using a simple grid search
for i in range(10,20):
    for j in range(150,200,5):

            circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT,i/10.0,j,
                                        param1=50,param2=30,minRadius=10,maxRadius=110)
            circles = np.round(circles[0, :]).astype("int")
            
            # ensure at least some circles were found
            if len(circles)>7 and len(circles)<10:
                output = image.copy()                
                # loop over the (x, y) coordinates and radius of the circles
                for (x, y, r) in circles:
                    # draw the circle in the output image, then draw a rectangle
                    # corresponding to the center of the circle
                    cv2.circle(output, (x, y), r, (0, 255, 0), 4)
                    cv2.rectangle(output, (x - 5, y - 5), (x + 5, y + 5), (0, 128, 255), -1)
                
                #print(f'{i} and {j}')
                #plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
                #plt.show()
                
```

### Coin class

```python
class Coin(object):
    
    def __init__(self, r,color, x=0,y=0):
        self.x=int(x)
        self.y=int(y)
        self.r=r
        self.color=color
                
        
def coin_metric(coin1 ,coin2):
        return abs(coin1.r-coin2.r)*10+ \
    (abs(coin1.color[0]-coin2.color[0])+abs(coin1.color[1]-coin2.color[1])+abs(coin1.color[2]-coin2.color[2]))
```

```python
#best match seems to be for parameter 1.9 and 175
circles = cv2.HoughCircles(gray, cv2.HOUGH_GRADIENT,1.9,175,param1=50,param2=30,minRadius=10,maxRadius=110)
circles = np.round(circles[0, :]).astype("int")

output = image.copy() 
height,width,depth = output.shape

results=[]
n=1
# loop over the (x, y) coordinates and radius of the circles
for (x, y, r) in circles:
    # draw the circle in the output image, then draw a rectangle
    # corresponding to the center of the circle
    cv2.circle(output, (x, y), r, (0, 255, 0), 4)
    cv2.rectangle(output, (x - 5, y - 5), (x + 5, y + 5), (0, 128, 255), -1)
    cv2.putText(output, str(n), (x,y), 5,4,(0, 255, 0), thickness=10) 
    n+=1
    
    
    #extracting the mean color of the circles
    circle_img = np.zeros((height,width), np.uint8)
    cv2.circle(circle_img, (x, y), r, (255, 255, 255), -1)
    color=cv2.mean(output, mask=circle_img)[::-1][1:4]
    results.append(Coin(x=x,y=y,r=r/(68.0/18.0),color=color))
    
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.show()
```

### Defining the actual coins

```python
# coins in the form {value:(diameter in mm, (color)}
british_coins={1:Coin(r=20.3,color=(104,102,65)), 2:Coin(r=25.9,color=(104,102,65)),\
               5:Coin(r=18.0,color=(85, 108, 73)), 10:Coin(r=24.5,color=(85, 108, 73)),\
               50:Coin(r=27.3,color=(85, 108, 73))} 
```

```python
def identify_coin(coin):
    min=np.inf
    coin_value=0
    for key, value in british_coins.items():
        if coin_metric(coin, value)<min:
            min=coin_metric(coin, value)
            coin_value=key
    return coin_value    
```

# The result

```python
output = image.copy() 
height,width,depth = output.shape
sum_=0
for coin in results:

        # calculate the coin value and draw it on the picture
        coin_value=identify_coin(coin)
        sum_+=coin_value
        cv2.putText(output, str(coin_value)+'p', (coin.x-35,coin.y+10), 5,3,(0, 0, 0), thickness=5) 
        
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.show()

#total sum on the picture
print(f'The sum of all coins is {sum_} pence.')
```
